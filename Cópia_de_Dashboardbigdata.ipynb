```python
# Agrupa por jogador e soma os gols
goleadores = df_players[df_players['Gols'] > 0].groupby('Player Name')['Gols'].sum().reset_index()
goleadores = goleadores.sort_values(by='Gols', ascending=False).head(20)
fig7 = px.bar(
        goleadores,
        y='Player Name',
        x='Gols',
        orientation='h',
        labels={'Total de Gols': 'Gols', 'Player Name': 'Nome do Jogador'},
        title='Top 20 Maiores Goleadores da Copa do Mundo',
        # color='Total de Gols',
        # color_continuous_scale='Viridis'  # Escala de cores para destacar os maiores goleadores
    )
fig7.show()
```

JOGADORES COM MAIS CARTÕES VERMELHOS

```python
cartoes = df_players[df_players['Event'] == 'R']
jogadoresavermelhados = cartoes['Player Name'].value_counts().reset_index()
jogadoresavermelhados.columns = ['Player Name','Red']
figure6 = px.bar(jogadoresavermelhados,x='Player Name',y='Red',title='Jogadores com mais cartões vermelhos',orientation='h')
figure6.show()
```

GRÁFICO TESTE

```python
# Filtra apenas as linhas com eventos que contenham gols
gols = df_players[df_players['Event'].str.contains(r'\bG\d+', na=False)]

# Agrupa os jogadores que mais fizeram gols contra
jogadorgols = gols.groupby('Player Name').size().reset_index(name='Total de Gols')

# Ordena do maior para o menor
jogadorgols = jogadorgols.sort_values(by='Total de Gols', ascending=False).head(20)

# Remove jogadores sem nome (caso existam)
jogadorgols = jogadorgols.dropna(subset=['Player Name'])

fig2 = px.bar(jogadorgols,
              x='Total de Gols',
              y='Player Name',
              orientation='h',
              title='Jogadores com mais gols',
              text='Total de Gols')

fig2.update_traces(textposition='outside')
fig2.update_layout(yaxis=dict(tickfont=dict(size=10)))
fig2.show()
```

COPAS COM MAIS GOLS

```python
df_cups['Year'] = df_cups['Year'].astype(str)
figure3 = px.bar(df_cups,x='Year',y='GoalsScored',title='Copas com mais gols',text='GoalsScored',labels={'Year':'Ano','GoalsScored':'Gols Feitos'}, hover_data=['MatchesPlayed'])
figure3.show()
```
